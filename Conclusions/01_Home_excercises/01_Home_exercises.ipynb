### ðŸ“š Exercise 1: Working with Numbers

Write a Python program that does the following:

1. Use the `input` function to prompt the user for two numbers: `a` and `b`
2. Compute:
   - The **sum** of `a` and `b`
   - The **product** of `a` and `b`
   - The **average** of `a` and `b`
3. Display the result of the operations with one decimal

```python
# Welcome message
print('This program divides two numbers.\n')

# Prompt user for input and convert to integer
a = float(input('Enter the first number: '))
b = float(input('Enter the second number: '))

# Calculate sum, product, and average
sum_ab = round(a + b, 1)
product_ab = round(a * b, 1)
average_ab = round(sum_ab / 2, 1)

# Display result with two decimals
print(f'\nSum = {sum_ab}')
print(f'Product = {product_ab}')
print(f'Average = {average_ab}')
#print(f'Average = {round(sum_ab/2, 1)}') # or pass the calculation directly to the f-string
```

### ðŸ“š Exercise 3: Temperature conversion program

Write a Python program that converts a temperature from Fahrenheit to Celsius.

The program should do the following:
1. Use the `input` function to prompt the user for a temperature in Fahrenheit
2. Convert the temperature to Celsius using the formula: $C = \left(\frac{5}{9}\right) \times (F - 32)$
3. Display the converted temperature rounded to the nearest integer

```python
# Print welcome message
print('*'*40)
print('**** Temperature Conversion Program ****')
print('*'*40)
print('This program converts from Fahrenheit to Celsius\n')

# Prompt user for input and convert to float
fahren = float(input('Enter temperature to convert: '))

# Calculate temperature conversion
celsius = (fahren-32) * 5 / 9
    
# Print conversion
print(f'\n{fahren} degree(s) Fahrenheit is {celsius:.0f} degree(s) Celsius.')
```
