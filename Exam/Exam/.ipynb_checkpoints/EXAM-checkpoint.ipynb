# Task 4. Add Region Information

```python
wildfires.isna().sum()
```

```python
#Filter on regional data from 2023 to see
regional_information = wildfires[wildfires['Year'] == 2023]
regional_information
```

```python
#Fill missing values
wildfires = wildfires.bfill()
```

```python
wildfires.isna().sum()
```

```python
#Aggregate total number of wildfires per region per year
total_wildfires_by_region = wildfires.groupby(['World regions according to OWID', 'Year']).agg(
    total_wildfires = ('Annual number of fires', 'sum'),
    year = ('Year', 'min'),  #Added column "year" to filter for 2012
   region = ('World regions according to OWID', 'max') #Added column region for plot in task 5
)

wildfires_df = pd.DataFrame(total_wildfires_by_region)
wildfires_df
```

I couldnt work out how to filter Year as it is not a column, so i added another column with year and region for filtration

```python
#Make a new DF for year 2012
wildfires_2012 = total_wildfires_by_region[total_wildfires_by_region['year'] == 2012].copy()
wildfires_2012.sort_values('total_wildfires', inplace = True)
wildfires_2012
```

# Task 5. Plot Regional Wildfires

```python
def wildfires_by_region(year):

    wildfire_year = total_wildfires_by_region[total_wildfires_by_region['year'] == year]
    wildfire_data = wildfire_year['total_wildfires'].sort_values(ascending = False)

    wildfire_region = total_wildfires_by_region['region'].unique()

    fig, ax = plt.subplots(figsize = (10, 5))

    ax.bar(
        wildfire_region,
        wildfire_data
    )

    ax.set_xlabel('Region')
    ax.set_ylabel('Annual wildfires')
    ax.set_title(f'Annual number of wildfires per region in year {year}')

wildfires_by_region(2015)
```

Additional plot for 2015:

```python
total_wildfires_2015 = wildfires.groupby('World regions according to OWID')['Annual number of fires'].sum()
total_wildfires_2015
```

```python
fig, ax = plt.subplots()

wildfire_region = total_wildfires_by_region['region'].unique()

ax.bar(
    wildfire_region,
    total_wildfires_2015.sort_values(ascending = False)
)

ax.set_xlabel('Region')
ax.set_ylabel('Number of wildfires')
ax.set_title('Number of wildfires per region, 2015')

plt.show()
```

# Task 7. Temperature Anomaly Calculation

```python
#Convert to datetime for calculation
temperatures['Day'] = pd.to_datetime(temperatures['Day'])
temperatures['Year'] = temperatures['Day'].dt.year
temperatures.head()
```

```python
#Filter new DF for data between 1981 to 2010
from1981to2010 = temperatures[(temperatures['Day'] > '1981-01-01') & (temperatures['Day'] < '2010-01-01')].copy()
```

```python
#Add column for year
from1981to2010.head()
```

```python
#Calculate average temperature between 1981 and 2010
avg_temp = from1981to2010.groupby(['Year', 'Entity'])['Monthly mean temperature'].mean()
avg_temp = pd.DataFrame(avg_temp)
avg_temp
```

```python
annual_mean_temp = temperatures.groupby(['Year', 'Entity'])['Monthly mean temperature'].mean()
annual_mean_temp = pd.DataFrame(annual_mean_temp)
annual_mean_temp
```

```python
anomaly = annual_mean_temp['Monthly mean temperature'] - avg_temp['Monthly mean temperature']
```

```python
anomaly
```
