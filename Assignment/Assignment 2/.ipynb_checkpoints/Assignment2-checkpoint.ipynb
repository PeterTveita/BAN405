Task 1

```python
import pandas as pd
```

Original dataframe:

```python
#1.1)
covid_df = pd.read_csv("data/time_series_covid19_deaths_global.csv") #Imports file to dataframe

covid_df #Original dataframe
```

Dataframe reshaped from wide to long format:

```python
#1.2)
#Pivots table from wide to long, with dates in a single column
melted_df = covid_df.melt(
    id_vars = ['Province/State', 'Country/Region', 'Lat', 'Long'],
    var_name = 'Date',
    value_name = 'Deaths'
)

melted_df #Melted dataframe
```

Converted date to datetime:

```python
#1.3)
melted_df['Date'] = pd.to_datetime(melted_df['Date'], format = '%m/%d/%y') #Converts 'Date' from object to datetime

melted_df
```

Aggregating deaths:

```python
#1.4)
deaths_country = melted_df.groupby(['Country/Region', 'Date'])['Deaths'].sum().reset_index() #Aggregates number of deaths per country per date

deaths_country['Date'] = pd.to_datetime(deaths_country['Date'], format = '%m/%d/%y') #Groupby changes 'Date' from datetime back to object

deaths_country #Dataframe for deaths
```

Calculating new deaths per day:

```python
#1.5)
deaths_country = deaths_country.sort_values(by = ['Country/Region', 'Date']) #Sort by country and date to get correct number of deaths

deaths_country['New deaths per day'] = deaths_country.groupby('Country/Region')['Deaths'].diff().fillna(0).astype(int) #Calculate diff for each line of deaths in each country.

deaths_country[deaths_country['New deaths per day'] != 0].head(10)
```

Plotting deaths for top 3 countries:

```python
#2.1)
import matplotlib.pyplot as plt

highest_deaths = deaths_country.groupby('Country/Region')['Deaths'].max().nlargest(3) #Filters the 3 countries with the most amounts of deaths
countries = highest_deaths.index.tolist() #Makes a list of the indexes of the series (country names)
data = deaths_country[deaths_country['Country/Region'].isin(countries)] #Makes a df with data from top 3 countries

fig, ax = plt.subplots(figsize = (12, 6))

#Plots a line for each of the top 3 countries
for country in countries:
    country_data = data[data['Country/Region'] == country]
    ax.plot(country_data['Date'], country_data['Deaths'], label = country)

#Formatting
ax.set_title('Deaths for top 3 countries')
ax.set_xlabel('Date')
ax.set_ylabel('Deaths')
ax.legend()
ax.ticklabel_format(style = 'plain', axis = 'y')

plt.savefig('total_deaths.png')
```

Subplots for Norway, Denmark and Sweden:

```python
#2.2)
NorDenSwe = ['Norway', 'Denmark', 'Sweden'] #List of countries

fig, axes = plt.subplots(3, 1, figsize = (8, 12))

for i, country in enumerate(NorDenSwe): #For each country, make a subplot, with the following:
    subset = deaths_country[deaths_country['Country/Region'] == country]

    axes[i].plot(subset['Date'], subset['New deaths per day'])
    axes[i].set_title(country)
    axes[i].set_ylabel('Deaths')
    axes[i].set_xlabel('Date')
    axes[i].set_ylim(0)
    axes[i].tick_params(axis = 'x', rotation = 45)

plt.tight_layout() #Adjust space between plots

#Save
plt.savefig('new_deaths.png')
plt.show()
```

Merging dataframes:

```python
#3.1)
continents = pd.read_csv('data/Countries Continents.csv')
continents

uniqueCont = continents['Entity'].unique() #Find unique contries in continents
#print(uniqueCont) #Made this a comment, so it doesnt print anymore
```

```python
diff = set(deaths_country['Country/Region']) - set(continents['Entity']) #Check which values dont match
diff

#Replaces values which are different to same value, and removes non-countries. This also makes diff 0, as there are no differences anymore
deaths_country['Country/Region'] = deaths_country['Country/Region'].replace({ 
    'US' : 'United States',
    'Korea, South' : 'South Korea',
    'Korea, North' : 'North Korea',
    'Burma' : 'Myanmar',
    'Cabo Verde' : 'Cape Verde',
    'Congo (Brazzaville)' : 'Republic of the Congo',
    'Congo (Kinsasha)' : 'Democratic Republic of the Congo',
    'Holy See' : 'Vatican City',
    'Taiwan*' : 'Taiwan',
    'Timor-Leste' : 'Timor',
    'West Bank and Gaza' : 'Palestine',
    'Micronesia': 'Micronesia (country)',
    'Diamond Princess' : None,
    'MS Zaandam' : None,
    'Summer Olympics 2020' : None,
    'Winter Olympics 2022' : None
})

deaths_country = deaths_country[deaths_country['Country/Region'].notna()] #Removes entities set to  "None"
```

```python
#Merges the two dataframes, and combines country-columns
merged_df = deaths_country.merge(
    continents,
    left_on = 'Country/Region',
    right_on = 'Entity',
    how = 'left'
)

merged_df
```

Left-joining dataframes, and verifying data:

```python
#3.2)
print(merged_df['Countries Continents'].value_counts(dropna = False)) #Displays amount of times each unique value occurs, including NaN-values

print("\nAmount of countries missing continent data, which countries, and how many observations this accounts for:\n")
print(merged_df.loc[merged_df['Countries Continents'].isna(), 'Country/Region'].nunique()) #Displays number of unique countries where value of 'Countries Continents' is NaN
print(merged_df.loc[merged_df['Countries Continents'].isna(), 'Country/Region'].unique()) #Displays unique countries
```

Calculating deaths per continent:

```python
#3.3)
#Deaths per continent
total_deaths = merged_df.groupby('Countries Continents')['Deaths'].max() #Finds continents with max cumulative deaths
print(total_deaths)
print("\nAs we can see, the North America was the continent with the most amount of deaths, with 1 123 836 deahts")
```

Plotting deaths per continent to bar plot:

```python
#3.4)
#Creating bar plot
plt.figure(figsize = (10, 5))
plt.bar(total_deaths.index, total_deaths.values)

#Formatting
plt.title('Total deaths per continent')
plt.xlabel('Continents')
plt.ylabel('Deaths')
plt.ticklabel_format(style='plain', axis='y')

#Saving
plt.savefig('continent_bar_plot.png')
plt.show()
```
